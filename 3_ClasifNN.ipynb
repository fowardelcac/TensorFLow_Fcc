```python
import tensorflow as tf
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler
```

```python
iris = load_iris()
iris.feature_names, iris.target_names
```

```python
scaler = MinMaxScaler()
```

```python
df_scaled = scaler.fit_transform(iris.data)
```

```python
df = pd.DataFrame(df_scaled, columns = iris.feature_names)
df.head()
```

```python
X_train_full, X_test, y_train_full, y_test = train_test_split(df, iris.target, test_size=0.2, random_state=42)
X_train, X_valid, y_train, y_valid_test = train_test_split(X_train_full, y_train_full, test_size=0.2, random_state=42)
```

```python
from tensorflow import keras
```

```python
model = keras.models.Sequential([
    keras.layers.Dense(4, input_shape = X_train.shape[1:]),
    keras.layers.Dense(300),
    keras.layers.Dense(3, activation = 'softmax')
])

model.compile(loss = 'sparse_categorical_crossentropy',
              optimizer = 'sgd',
              metrics = ['accuracy'])
model.fit(X_train, y_train, epochs=32, validation_data = (X_valid, y_valid_test))
```

```python
from keras import layers, models, optimizers

model = models.Sequential([
    layers.Dense(4, input_shape=X_train.shape[1:]),
    layers.Dense(500, activation='relu'),  # Cambio 1: Aumento del número de neuronas
    layers.Dropout(0.5),  # Cambio 2: Agregar dropout para regularización
    layers.Dense(3, activation='softmax')
])

# Cambio 3: Cambiar el optimizador y ajustar la tasa de aprendizaje
optimizer = optimizers.Adam(learning_rate = 0.001)

model.compile(loss='sparse_categorical_crossentropy',
              optimizer=optimizer,
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=50, validation_data=(X_valid, y_valid_test))
```

```python
# Evaluating the model on the training set
train_loss = model.evaluate(X_train, y_train)

# Evaluating the model on the validation set
val_loss = model.evaluate(X_valid, y_valid_test)

# Print the evaluation results
print("Training Loss:", train_loss)
print("Validation Loss:", val_loss)
```

```python
test_loss = model.evaluate(X_test, y_test)

# Print the test loss
print("Test Loss:", test_loss)
```

```python
y_label = np.array(y_test[:40]).reshape(-1, 1)

X_new_scaled = scaler.fit_transform(X_test[:40])

# Make predictions using the preprocessed data
predictions = model.predict(X_new_scaled)
df = pd.DataFrame(predictions, columns = ['setosa_pred', 'versicolor_pred', 'virginica_pred'])
df['y_label'] = y_label
```
